```python
from google.cloud import bigquery

# 1. Cria o cliente do BigQuery usando o arquivo de credenciais
client = bigquery.Client.from_service_account_json('/tmp/google_creds.json')

# 2. Query de Extração
query = """
SELECT 
    ano_campeonato, 
    rodada, 
    data, 
    time_mandante, 
    time_visitante, 
    gols_mandante, 
    gols_visitante 
FROM `basedosdados.mundo_transfermarkt_competicoes.brasileirao_serie_a` 
WHERE ano_campeonato = 2023
ORDER BY data, rodada ASC
"""

print("Extraindo dados do BigQuery...")

# 3. Executa a query e converte para Pandas
query_job = client.query(query)
df_pandas = query_job.to_dataframe()

# 4. Conversão para Spark e Exibição
df_spark = spark.createDataFrame(df_pandas)
display(df_spark)
```
